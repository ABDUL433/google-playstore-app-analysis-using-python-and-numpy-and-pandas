```python
import pandas as pd
data=pd.read_csv('googleplaystore.csv')
data

```

```python
Q1=data['Rating'].quantile(0.25)
Q3=data['Rating'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q1+1.5*IQR
```

```python
filterdata=data[(data['Rating']>=lower_bound)&(data['Rating']<=upper_bound)]
```

```python
 import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.boxplot(x='Rating',data=filterdata,color='red')
plt.title('boxplot of App Ratings')
plt.xlabel('Rating')

plt.show()
```

```python
def convert_size(size):
    ''' This function will convert the size column to numeric by multiplying the values with 1024
    if it has 'K' in it and 1024*1024 if it has 'M' in it.'''
    if isinstance(size,str):
        if 'k' in size:
          return float(size.replace('k','')) * 1024
    elif 'M' in size:
        return float(size.replace('M',''))*1024*1024
    elif 'Varies with device' in size:
        return np.nan
        return (size)
        
        
    

     

    

    
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('googleplaystore.csv')
```

```python
df.head()
```

```python
df['Installs']=df['Installs'].astype(str)
df['Installs']=df['Installs'].apply(lambda x:x.replace('Free','0')if 'Free' in x else x)
df['Installs']=df['Installs'].apply(lambda x:x.replace('+',''))
df['Installs']=df['Installs'].str.replace(',','')
df['Installs']=df['Installs'].astype(int)
```

```python
df['Installs']=df['Installs'].apply(lambda x:x.replace('+','') if '+' in str(x) else x)
df['Installs']=df['Installs'].apply(lambda x:x.replace(',','') if '.' in str(x) else x)
df['installs']= df['Installs'].apply( lambda x:int(x))                               
                                    
```
